* To have a first quick look on the data, check for outliers or evident patterns and check for outputs of multifactor analysis
* To have a finer look on model errors. 

We'll be going through the very basics of `matplotlib` which is one of the first visualisation APIs in `python`. Most modern ones are wrappers over `matplotlib`.

For this first lecture we'll just be looking into plotting maths functions, scatter plots and histograms. 

## Plotting Maths Functions

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

When you're in a notebook there are two ways to show a plot, either as a *static* png or as interactive plots. We'll be only taking things statically, you're welcome to try out the interactive version. For interactivity one uses the magic command `%matplotlib notebook` once per session.

```python
%matplotlib inline
```

There are two approaches to the `matplotlib` syntax. A `MATLAB` historical one and a more object oriented pythonic one. We'll be sticking to the latter during this lecture. It's more adapted to your background.

```python
x = np.linspace(0, 10, 1000)  # generating a regular sequence of real numbers

fig, ax = plt.subplots(figsize=(10, 8)) 
# An object generating a figure (one unique per session) 
# and axes (where you draw stuff).

ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x), '--')
```

It is an easy feat using this syntax to draw out things in multiple plots.

```python
fig_a, ax_a = plt.subplots(2, figsize=(10, 8))

ax_a[0].plot(x, np.sin(x))
ax_a[1].plot(x, np.cos(x), '--')
```

## Simple Scatter Plots

```python
x = np.random.randn(1000)
y = np.random.randn(1000)

fig_b, ax_b = plt.subplots(figsize=(10, 8))

ax_b.scatter(x, y)
```

Equivalent result can be given using the `plot` function.

```python
fig_c, ax_c = plt.subplots(figsize=(10, 8))

ax_c.plot(x, y, 'o')
```

## Histograms

```python
for_hist = np.random.randn(2000)

fig_d, ax_d = plt.subplots(figsize=(10, 8))
ax_d.hist(for_hist, bins=30, edgecolor='black')
```
